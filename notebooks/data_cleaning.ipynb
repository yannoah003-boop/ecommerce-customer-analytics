```python
df=df.dropna(subset=['CustomerID','Description'])
df=df[(df['UnitPrice']>0) & (df['Quantity']>0) & (df['UnitPrice']<1000)].copy()
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
df['CustomerID']=df['CustomerID'].astype(int)
df=df.copy()
df.shape
```

```python
df['Revenue']=(df['Quantity']*df['UnitPrice']).round(2)
df[['Quantity','UnitPrice','Revenue']].head()
```

```python
df['YearMonth']=df['InvoiceDate'].dt.to_period('M')
df['YearMonth']=df['YearMonth'].astype(str)
df['Date']=df['InvoiceDate'].dt.date
```

```python
print('rows,cols:',df.shape)
print('null check:/n',df.isnull().sum())
print('min Quantity:',df['Quantity'].min())
print('UnitPrice Range:',df['UnitPrice'].min(),'to',df['UnitPrice'].max())
print('unique Customers:',df['CustomerID'].nunique())
print('date range:',df['InvoiceDate'].min(),'to',df['InvoiceDate'].max())
```
